```python
import numpy as np
import pandas as pd
```

**Load Data**

```python
df=pd.read_csv("/kaggle/input/datasets/carrie1/ecommerce-data/data.csv",encoding="latin1")

print(df)
```

**Missing value Check**

```python
print(df.isnull())
```

```python
df.isnull().sum()
```

**Data Cleaning**

```python
# missing value remove
df = df.dropna()

# duplicate remove
df = df.drop_duplicates()

# date convert
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.isnull().sum()
```

**Total Revenue**

```python
df['TotalPrice'] = np.multiply(df['Quantity'], df['UnitPrice'])

print(df)
```

**TotalPrice**

```python
total=df["TotalPrice"].sum()
print(total)
```

**Monthly Sales**

```python
df['Month'] = df['InvoiceDate'].dt.month

monthly_sales = df.groupby('Month')['TotalPrice'].sum().reset_index()

print("Monthly Sales:")
print(monthly_sales.sort_values(by='TotalPrice', ascending=False))
```

**Top Products**

```python
top_products = df.groupby('Description')['TotalPrice'].sum().reset_index()

top_products = top_products.sort_values(by='TotalPrice', ascending=False)

print("Top Products:")
print(top_products.head(10))
```

**Top Customers**

```python
top_customers = df.groupby('CustomerID')['TotalPrice'].sum().reset_index()

top_customers = top_customers.sort_values(by='TotalPrice', ascending=False)

print("Top Customers:")
print(top_customers.head(10))
```

**Country Analysis**

```python
country_sales = df.groupby('Country')['TotalPrice'].sum().reset_index()

country_sales = country_sales.sort_values(by='TotalPrice', ascending=False)

print("Top Countries:")
print(country_sales.head(10))
```

**Repeat Customers**

```python
repeat_customers = df.groupby('CustomerID')['InvoiceNo'].nunique()

repeat_customers = repeat_customers[repeat_customers > 1]

print("Repeat Customers Count:", repeat_customers.count())
```
